# Zad. 2
Utwórz nową kolumną o nazwie <tt>HasCabin</tt>, która zawiera 1 jeżeli ktoś miał swoją kabinę i 0 jeśli nie.

```python
noCabin = df['Cabin'].isnull()
df['HasCabin'] = 1
df.loc[noCabin, 'HasCabin'] = 0
df.head(20)

# df['HasCabin'] = np.nan
# df.loc[df.Cabin.notnull(), 'HasCabin'] = 1
# df.loc[df.Cabin.isnull(), 'HasCabin'] = 0
# df = df.drop(['Cabin'], axis=1)
```

```python
# nie było w poleceniu ale usuwanie kolumny oryginalnej -> Cabin
df.drop(axis='cols', columns=['Cabin'], inplace=True)
df.head()
```

# Zad. 3
Usuń brakujące wartości

```python
# usuwanie wierszy zawierających NaN
df.dropna(inplace=True)
df.head()
```

# Zad. 4
Dokonaj konwersji danych do postaci numerycznej

```python
nominal_columns = ['Sex', 'Embarked']
dummy_df = pd.get_dummies(df[nominal_columns])
dummy_df.head()
```

```python
# dodanie kolumn dummy_df do df
df = pd.concat([df, dummy_df], axis=1)
df.head()
```

```python
df.drop(columns=['Sex', 'Embarked'], inplace=True)
```
