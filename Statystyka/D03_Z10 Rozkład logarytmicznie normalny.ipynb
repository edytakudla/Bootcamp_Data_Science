# Rozkład logarytmicznie normalny

Rozkład logarytmicznie normalny ma gęstość

$$
LN(x;\mu,\sigma^2) = 
\left\{ \begin{array}{ll}
{\displaystyle {\frac {1}{\sigma {\sqrt {2\pi }}}}\ e^{-{\frac {\left(\ln x-\mu \right)^{2}}{2\sigma ^{2}}}}} & \textrm{gdy $x \geq 0$}\\
0 & \textrm{gdy $x < 0$}
\end{array} \right. .
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

Definiujemy zmienną losową

```python
mi, sigma = 1,1
x = np.arange(0,10, 0.05)
lognorm = stats.lognorm(mi, sigma)
lognorm
```

Narysujmy gęstość zmiennej losowej

```python
plt.plot(x, lognorm.pdf(x), '--', ms=5, label = 'PDF(X)')
plt.legend()
plt.show()
```

Narysujmy dystrybuantę zmiennej losowej

```python
plt.plot(x, lognorm.cdf(x), '--', ms=5, label='CDF(x)')
plt.legend()
plt.show()
```

Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
trials = lognorm.rvs(1000)
fig, ax = plt.subplots(1,1)
ax.plot(x, lognorm.pdf(x), 'k--', lw=2, label = 'mi = 1, sigma = 1')
ax.hist(trials, density=True, label = 'histogram')
plt.legend()
plt.show()
```

Narysujmy gęstości rozkładu logarytmicznie normalnego z różnymi parametrami

```python
x = np.arange(0,10, 0.05)
fig, ax = plt.subplots(1,1)

mi1, sigma1 = 3,5
lognorm1 = stats.lognorm(mi1, sigma1)

mi2, sigma2 = 1,1
lognorm2 = stats.lognorm(mi2, sigma2)

mi3, sigma3 = 2,3
lognorm3 = stats.lognorm(mi3, sigma3)

ax.plot(x, lognorm1.pdf(x), 'b--', lw=2, label = 'mi = 3, sigma = 5')
ax.plot(x, lognorm2.pdf(x), 'r--', lw=2, label = 'mi = 1, sigma = 1')
ax.plot(x, lognorm3.pdf(x), 'k--', lw=2, label = 'mi = 2, sigma = 3')

plt.legend()
plt.show()
```

Narysujmy kilka histogramów różnych próbek z rozkładu logarytmicznie normalnego o tych samych parametrach.

```python
mi, sigma = 1,1
lognorm = stats.lognorm(mi, sigma)

fig, ax = plt.subplots(5,5, figsize=(10,10))
for i in np.arange(5):
    for j in np.arange(5):
        trials = lognorm.rvs(100)
        ax[i,j].hist(trials, density=True)
        
plt.show()
```
