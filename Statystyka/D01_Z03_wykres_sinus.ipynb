Napiszmy skrypt w Python, który wygeneruje punkty w kształcie sinusa i je wyrysuje.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange(-10,10,1)
plt.plot(x, x**2, 'bo')
```

```python
MIN = -8 * np.pi
MAX = 8 * np.pi
STEP = 0.1
x = np.arange(MIN, MAX, STEP)
plt.plot(x, np.sin(x), 'r>')
plt.xlabel('x')
plt.ylabel('sin(x)')
```
