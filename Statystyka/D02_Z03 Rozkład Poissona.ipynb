# Rozkład Poissona
Rozkład prawdopodobieństwa zmiennej losowej $X$ o rozkładzie Poissona z parametrem $\lambda > 0$ wyraża się wzorem:
$$
P(X = k) = e^{-\lambda} \cdot \frac{\lambda^{k}}{k!}.
$$


```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

Zdefiniuj zmienną losową o rozkładzie Poissona  $\lambda=2$

```python
lam = 2

poissonDist = stats.poisson(lam)
print(poissonDist)
```

Narysuj gęstość zmiennej losowej o rozkładzie Poissona

```python
lam = 2
fig, ax = plt.subplots(1,1)
x = np.arange(35)
print(stats.poisson.pmf(x, lam))
ax.plot(x, stats.poisson.pmf(x, lam), 'cp--')
plt.show()
```

```python
# rozwiązanie Przemka
fig, ax = plt.subplots(1, 1)
ax.set_xlim([-1,20])
ax.set_ylim([0,0.3])
x = np.arange(0, 35)

ax.plot(x, poissonDist.pmf(x), 'bo', ms=8, label='poisson pmf')

ax.vlines(x, 0, poissonDist.pmf(x), colors='b', lw=5, alpha=0.5)
ax.legend()
plt.show()
```

Wylosuj próbkę i narysuj histogram

```python
trials = poissonDist.rvs(10000)
# print(trials)
plt.hist(trials, bins=20)
plt.show()
```

Narysuj dystrybuantę zmiennej losowej o rozkładzie Poissona

```python
fig, ax = plt.subplots(1,1)
x = np.arange(-2, 10, 0.01)
lam = 2
print(stats.poisson.cdf(x, lam))
ax.plot(x, stats.poisson.cdf(x, lam), 'yh-')
plt.show()
```

```python
# rozwiązanie Przemka
p = 0.7
fig, ax = plt.subplots(1, 1)
ax.set_xlim([-2,10])
ax.set_ylim([-0.2,1.2])

x = np.arange(-2, 10,0.01)
ax.plot(x, stats.poisson.cdf(x, lam), 'bo--', ms=5, label='poisson cdf')
plt.legend()
plt.show()
```

Policz momenty

```python
lam = 5
mean, var, skew, kurt = stats.poisson.stats(lam, moments='mvsk')
print(mean)
print(var)
print(skew)
print(kurt)
```

Narysuj gęstość zmiennej losowej dla różnych parametrów

```python
fig, ax = plt.subplots(1,1)

x = np.arange(20)
ax.plot(x, stats.poisson.pmf(x, 1), 'g>-', ms=6, label='lam=1, n=20')
ax.plot(x, stats.poisson.pmf(x, 2), 'yh-', ms=6, label='lam=2, n=20')
ax.plot(x, stats.poisson.pmf(x, 5), 'bo-', ms=6, label='lam=5, n=20')
ax.plot(x, stats.poisson.pmf(x, 3), 'r<-', ms=6, label='lam=3, n=20')
plt.legend()
plt.show()
```

```python
# rozwiązanie Przemka
# Arbitrarily select 3 lambda values
lambdas = [1,4,5,10]

k = np.arange(20)       # generate x-values
markersize = 8
for par in lambdas:
    plt.plot(k, stats.poisson.pmf(k, par), 'o--',
             label='$\lambda={0}$'.format(par))

# Format the plot
plt.legend()
plt.title('Poisson distribution')
plt.xlabel('X')
plt.ylabel('P(X)')
plt.show()
```
