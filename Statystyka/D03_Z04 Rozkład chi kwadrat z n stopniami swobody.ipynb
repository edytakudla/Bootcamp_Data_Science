# Rozkład chi kwadrat z $n$ stopniami swobody
Rozkład chi kwadrat ma gęstość:

$$
f_{X}(x) = 
\left\{ \begin{array}{ll}
{\frac {1}{2^{\frac {n}{2}}\Gamma \left({\frac {n}{2}}\right)}}\;x^{{\frac {n}{2}}-1}e^{-{\frac {x}{2}}} 
   & \textrm{gdy } x \geq 0\\
0 & \textrm{gdy } x < 0 
\end{array} \right. ,
$$

gdzie $\Gamma$ oznacza funkcję Gamma https://pl.wikipedia.org/wiki/Funkcja_\%CE\%93


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

# Definiujemy zmienną losową

```python
t = np.arange(0, 20, 0.05)
n=4
X=stats.chi2(n)
```

# Narysujmy gęstość zmiennej losowej

```python
plt.plot(t, X.pdf(t))
plt.show()
```

```python
# rozwiązanie Przemka - różne stopnie swobody
fig, ax = plt.subplots(1, 1)
plt.plot(t, stats.chi2(4).pdf(t), '--', ms=8, label='n=4')
plt.plot(t, stats.chi2(6).pdf(t), '--', ms=8, label='n=6')
plt.plot(t, stats.chi2(8).pdf(t), '--', ms=8, label='n=8')
plt.plot(t, stats.chi2(10).pdf(t), '--', ms=8, label='n=10')
plt.plot(t, stats.chi2(20).pdf(t), '--', ms=8, label='n=20')
plt.legend()
plt.show()
```

# Narysujmy dystrybuantę zmiennej losowej

```python
plt.plot(t, X.cdf(t))
plt.show()
```

# Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
trials = X.rvs(1000)

fig, ax = plt.subplots(1,1)
ax.plot(t, X.pdf(t), 'r--', label='PDF(X)')
ax.hist(trials, label = 'histogram', histtype='stepfilled')
plt.legend()
plt.show()
```

# Narysujmy gęstości rozkładu z różnymi parametrami

```python
fig, ax = plt.subplots(1, 1)
plt.plot(t, stats.chi2(1).pdf(t), ls="--",label='n=1')  
plt.plot(t, stats.chi2(2).pdf(t), ls="--",label='n=2')  
plt.plot(t, stats.chi2(3).pdf(t), ls="--",label='n=3')  
plt.plot(t, stats.chi2(4).pdf(t), ls="--",label='n=4')  
plt.legend()
plt.show()
```

# Narysujmy kilka histogramów różnych próbek z rozkładu o tych samych parametrach.

```python
fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        trials = X.rvs(size=100)
        ax[i,j].hist(trials, density=True)
plt.show()
```
