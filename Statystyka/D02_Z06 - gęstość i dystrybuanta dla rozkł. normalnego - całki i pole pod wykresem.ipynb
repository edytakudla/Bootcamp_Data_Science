# Gęstość i dystrybuanta rozkładu Normalnego

Jeżeli $f_{X} \colon \mathbb{R} \to [0,+\infty] $ jest gęstością ciągłej zmiennej losowej $X$ to
$$
\int_{-\infty}^x f_{X}(t)dt=P((-\infty, x])=F_X(x),
$$
gdzie $F_X$ jest dystrybuantą zmiennej losowej $X$. 


Jeżeli $f_{X} \colon \mathbb{R} \to [0,+\infty] $ jest gęstością ciągłej zmiennej losowej $X$ to
$$
\int_{-\infty}^x f_{X}(t)dt=P((-\infty, x])=F_X(x),
$$
gdzie $F_X$ jest dystrybuantą zmiennej losowej $X$. 

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy import integrate
```

Zdefiniujmy zmienną losową o rozkładzie Normalnym

```python
(mi, sigma)=(0,1)
gauss=stats.norm(mi, sigma)
```

Będziemy liczyć prawdopodobieństwo $P(X \in [-\infty,x_0] )$, gdzie $x_0 \in \mathbb{R}$

```python
x0=0.6
```

```python
# pdf - gęstość ciągłej zmiennej losowej
# integrate.quad - całka oznaczona 
(inte,err)=integrate.quad(lambda x: gauss.pdf(x), -np.inf, x0)
print('inte: ', inte)
print('err: ', err)
print('inte zaokr. do 4 miejsc: ', np.round(inte,4))
```

Zgodnie z naszymi wzorami
$$
P(X \in (-\infty,x_0] )  = \int_{\infty}^{x_{0}} f_{X}(x)dx=P(X\leq x_0)=F_{X}(x_0) 
$$

```python
# 100 liczb z przedziału [-3, 3]
x = np.linspace(-3,3,100)
# PDF - gęstość zmiennej losowej f(x)
# CDF - dystrybuanta zmiennej losowej F(x)
yp = gauss.pdf(x)
y = gauss.cdf(x)
x0 = 0.6
x1 = np.linspace(-3, x0)
y1 = gauss.pdf(x1)

figs, axs = plt.subplots(1,2, figsize=(10, 4))

axs[0].plot(x,yp, 'k')
axs[0].fill_between(x1, y1, facecolor='#CCCCCC')
axs[0].text(x0, 0.3, f'pole = {np.round(inte,4)}')
axs[0].set_xlabel('x')
axs[0].set_ylabel('GĘSTOŚĆ PDF(x)')

axs[1].plot(x, y, '#999999', lw=3)
axs[1].set_xlabel('x')
axs[1].set_ylabel('DYSTRYBUANTA CDF(x)')
axs[1].text(x0, gauss.cdf(x0), np.round(gauss.cdf(x0),4))
plt.vlines(x0, 0, gauss.cdf(x0), linestyles='--')
plt.show()
```

# Zadanie 2
Napisz skrypt, który będzie liczył prawdopodobieństwo:

$$
P(X \in [-\infty,x_1] \cup [x_2\infty])  = \int_{\infty}^{x_{1}} f_{X}(x)dx + \int_{x_{2}}^{\infty} f_{X}(x)dx=P(X\leq x_1 \ or \ X\geq x_2)=F_{X}(x_1)+1-F_{X}(x_2) 
$$
<ul>
<li>Sprawdź czy wszystkie równości zachodzą? </li>
<li>Narysuj gęstość i dystrybuantę oraz zaznacz na niej wszystkie dane. </li>
</ul>

```python
# 100 liczb z przedziału [-3, 3]
x = np.linspace(-3,3,100)
# PDF - gęstość zmiennej losowej f(x)
# CDF - dystrybuanta zmiennej losowej F(x)
y = gauss.cdf(x)
x01 = -0.5
x02 = 0.5

x11 = np.linspace(-3,x01)
y11 = gauss.pdf(x11)

x22 = np.linspace(x02,3)
y22 = gauss.pdf(x22)

figs, axs = plt.subplots(1,2, figsize=(10, 4))

axs[0].plot(x,gauss.pdf(x), 'k')
axs[0].fill_between(x11, y11, facecolor='#CCCCCC')
axs[0].fill_between(x22, y22, facecolor='#CCCCCC')
axs[0].text(x01, 0.1, f'pole1 = {np.round(inte1,4)}')
axs[0].text(x02, 0.3, f'pole2 = {np.round(inte2,4)}')
axs[0].set_xlabel('x')
axs[0].set_ylabel('GĘSTOŚĆ PDF(x)')

axs[1].plot(x, y, '#999999', lw=3)
axs[1].set_xlabel('x')
axs[1].set_ylabel('DYSTRYBUANTA CDF(x)')
axs[1].text(x01, gauss.cdf(x01), np.round(gauss.cdf(x01),4))
axs[1].text(x02, gauss.cdf(x02), np.round(gauss.cdf(x02),4))
plt.vlines(x01, 0, gauss.cdf(x01), linestyles='--')
plt.vlines(x02, 0, gauss.cdf(x02), linestyles='--')
plt.show()
```

```python
# sprawdzenie czy ok
gauss.cdf(x01)+1-gauss.cdf(x02)
```

```python
# integrate.quad - całka oznaczona 
(inte1,err1)=integrate.quad(lambda x: gauss.pdf(x), -np.inf, x01)
print('inte1: ', inte1)

(inte2,err2)=integrate.quad(lambda x: gauss.pdf(x), x02, np.inf)
print('inte2: ', inte2)

pole = inte1 + inte2
print('pole sumy: ', pole)
```
