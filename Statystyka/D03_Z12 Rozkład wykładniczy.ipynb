# Rozkład wykładniczy

Rozkład wykładniczy ma gęstość:

$$
f_{X}(x) = 
\left\{ \begin{array}{ll}
\frac{k}{\lambda} \lambda e^{-\lambda x} & \textrm{gdy $x \geq 0$}\\
0 & \textrm{gdy $x < 0$}
\end{array} \right. ,
$$

gdzie $k>0$ jest parametrem kształtu oraz $\lambda> 0$ jest parametrem skali.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

# Definiujemy zmienną losową

```python
lam = 1
x = np.arange(-5,5,0.1)
expon = stats.expon(lam)
expon
```

# Narysujmy gęstość zmiennej losowej

```python
plt.plot(x, expon.pdf(x), '--', ms=5, label = 'PDF(X)')
plt.legend()
plt.show()
```

# Narysujmy dystrybuantę zmiennej losowej

```python
plt.plot(x, expon.cdf(x), '--', ms=5, label = 'CDF(X)')
plt.legend()
plt.show()
```

# Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
trials = expon.rvs(1000)
fig, ax = plt.subplots(1,1)
ax.plot(x, expon.pdf(x), 'k--', lw=2, label = 'lam = 1')
ax.hist(trials, density=True, label = 'histogram')
plt.legend()
plt.show()
```

# Narysujmy gęstości rozkładu wykładniczego z różnymi parametrami

```python
x = np.arange(0,10, 0.05)
fig, ax = plt.subplots(1,1)

lam1 = 1
expon1 = stats.expon(lam1)

lam2 = 2
expon2 = stats.expon(lam2)

lam3 = 3
expon3 = stats.expon(lam3)

ax.plot(x, expon1.pdf(x), 'b--', lw=2, label = 'lam = 1')
ax.plot(x, expon2.pdf(x), 'r--', lw=2, label = 'lam = 2')
ax.plot(x, expon3.pdf(x), 'k--', lw=2, label = 'lam = 3')

plt.legend()
plt.show()
```

# Narysujmy kilka histogramów różnych próbek z rozkładu wykładniczego o tych samych parametrach.

```python
lam = 1
expon = stats.expon(lam)

fig, ax = plt.subplots(5,5, figsize=(10,10))
for i in np.arange(5):
    for j in np.arange(5):
        trials = expon.rvs(100)
        ax[i,j].hist(trials, density=True)
        
plt.show()
```
