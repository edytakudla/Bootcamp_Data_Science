# Rozkład normalny - GAUSSA
Gęstość zmiennej losowej $X$ o rozkładzie normalnym z parametrami $\mu$ i $\sigma$ jest dana przez 
$$
f(x) = \frac{1}{\sigma \sqrt{2 \pi} } e^{-\frac{(x-m)^2}{2\sigma^2} }.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

Definiujemy zmienną losową

```python
mi=0 
sigma=1
gauss = stats.norm(mi, sigma)
gauss
```

Narysujmy gęstość zmiennej losowej

```python
x = np.arange(-5,5, 0.01)
fig, ax = plt.subplots(1,1)
plt.plot(x, gauss.pdf(x), 'b--', ms=5, label='gęstość zmiennej losowej w rozkładzie normalnym / frozen pdf')
ax.set_xlim([-5,5])
ax.set_ylim([0,0.4])
plt.legend(loc='best')
plt.show()
```

Narysujmy dystrybuantę zmiennej losowej

```python
x = np.arange(-5,5, 0.01)
fig, ax = plt.subplots(1,1)
plt.plot(x, gauss.cdf(x), 'b--', ms=5, label='dystrybuanta zmiennej losowej w rozkładzie normalnym')
plt.legend(loc='best')
plt.show()
```

Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
x = np.arange(-5 ,5, 0.01)
trials = gauss.rvs(1000)
fig, ax = plt.subplots(1,1)
ax.hist(trials, density=True, alpha=0.5, label='histogram gauss')
ax.plot(x, gauss.pdf(x), 'k--', lw=2, label='mi=0, sigma=1')
plt.legend()
plt.show()
```

Narysujmy gęstości rozkładu normalnego z różnymi parametrami

```python
x = np.arange(-5 ,5, 0.01)
fig, ax = plt.subplots(1,1)

mi1=0 
sigma1=1
gauss1 = stats.norm(mi1, sigma1)

mi2=1 
sigma2=2
gauss2 = stats.norm(mi2, sigma2)

mi3=1 
sigma3=3
gauss3 = stats.norm(mi3, sigma3)

ax.plot(x, gauss1.pdf(x), 'k--', label='mi=0, sigma=1')
ax.plot(x, gauss2.pdf(x), 'k--', label='mi=1, sigma=2')
ax.plot(x, gauss3.pdf(x), 'k--', label='mi=1, sigma=3')
plt.show()
```

```python
# rozwiążanie Przemka
sigma_values = [0.5, 1.0, 2.0]
mi_values = [0, 1, 2]
fig, ax = plt.subplots(1, 1)
for mi, sigma in zip(mi_values, sigma_values):
    gauss = stats.norm(mi, sigma)
    plt.plot(x, gauss.pdf(x), ls="--", c='black', label=r'$\mu=%1.f,\ \sigma=%.1f$' % (mi, sigma))
ax.set_xlim([-5,5])
ax.set_ylim([-0.1,1.1])    
plt.legend()
plt.show()
```

Narysujmy kilka histogramów różnych próbek z rozkładu normalnego o tych samych parametrach.

```python
mi=0
sigma=1
gauss = stats.norm(mi, sigma)
trials1000 = gauss.rvs(1000)
trials100 = gauss.rvs(100)
trials500 = gauss.rvs(500)
plt.hist(trials1000, density=True, alpha=0.5)
plt.hist(trials100, density=True, alpha=0.5)
plt.hist(trials500, density=True, alpha=0.5)
plt.show()
```

```python
gauss=stats.norm(0,1)
fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        trials = gauss.rvs(size=100)
        ax[i,j].hist(trials, density=True)
plt.show()
```
