# LAB - S02-L009 - Standardowe metody Pythona a obiekty PANDAS

```python
# Z pliku StackOverflowDeveloperSurvey.csv wczytaj kolumnę Salary jako Data Series i zapisz ją w zmiennej
# salary. Wczytując elementy pomiń te ankiety, w których respondenci nie podali kwoty wynagrodzenia (użyj
# dropna()). Wyświetl 5 pierwszych elementów serii
salary = pd.read_csv('C:\PythonScripts\course-files\StackOverflowDeveloperSurvey.csv', usecols=['Salary'], squeeze=True).dropna()
salary.head()
```

```python
# Ile osób odpowiedziało w ankiecie na temat otrzymywanego wynagrodzenia?
len(salary)
```

```python
# Jakie jest najmniejsze wynagrodzenie? A jakie jest największe wynagrodzenie?
print(min(salary))
print(max(salary))
```

```python
# Pięć pierwszych elementów serii zamień na listę (nie zapisuj jej w zmiennej, po prostu wyświetl)
list(salary.head())
```

```python
# Pięć pierwszych elementów serii zamień na słownik (nie zapisuj go, po prostu wyświetl)
dict(salary.head())
```

```python
# W zmiennej listSalarySorted zapisz listę powstałą z posortowania serii salary malejąco
listSalarySorted = list(salary.sort_values(ascending=False))
listSalarySorted
# drugi sposob: listSalarySorted = sorted(salary,reverse=True)
```

```python
# Wyświetl pięć pierwszych elementów listy listSalarySorted. (Pamiętaj że to nie seria i nie zadziała metoda
# head()
listSalarySorted[0:6]
```
