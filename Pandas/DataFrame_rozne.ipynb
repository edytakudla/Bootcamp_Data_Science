## 1. Metoda dropna()

```python
mc_frame = pd.read_csv('C:\PythonScripts\course-files\Mcdonalds.csv')
mc_frame.head()
```

```python
# Ponieważ dane są zbyt "ładne" chcemy je trochę popsuć. Dodajemy w 2 miejsca wartość NaN, tak, jakbyśmy tych danych
# fakktycznie nie mieli.
mc_frame.loc[1, 'Calories'] = np.NaN
mc_frame.loc[2, 'TotalFat'] = np.NaN
```

```python
# Subset ozn. Listę kolumn na które polecenie drop ma popatrzeć aby podjąć decyzję czy wiersz ma być usunięty czy nie
mc_frame.dropna(how='any', subset=['Calories', 'TotalFat']).head()
```

```python
# Wyszukaj te firmy, które w kolumnie Industry zawierały napis comp. 
# Zadbaj o to aby odnalezione były dane niezależnie od tego czy w kolumnie tekst jest napisany małymi czy wielkimi literami.
comp = fortune['Industry'].str.lower().str.contains('comp')
fortune[comp].head(5)
```

```python
# Dodaj do fortune dwie nowe kolumny 'City','State', które powstaną przez rozbicie 
# kolumny Hqlocation ze względu na przecinek. Wyświetl nagłówek fortune
fortune[['City', 'State']] = fortune['Hqlocation'].str.split(',', expand=True)
fortune.head()
```

```python
# Chcemy do obiektu frame dodać kolumnę IndustryShort, która powstanie wskutek połączenia pierwszych liter wyrazów 
# znajdujących się w kolumnie Industry (nie przejmuj się nawiasami, przecinkami itp. W tym celu:

#     Stwórz funkcje BuildShortcut, która jako parametr przyjmie wiersz
#     Funkcja ma pobrać wartość z kolumny Industry i zapisać ją w zmiennej industry
#     Zadeklarować pusty napis result
#     Dla każdego napisu znajdującego się na liście powstałej w skutek rozbicia industry ze względu na spacje
#     Pobrać pierwszą literę słowa i dodać ją do napisu result
#     Zwrócić napis result

def Buildshortcut(row):
    industry = row['Industry']
    result = ''
    industry = industry.split(' ')
    for word in industry:
        result += word[0]
    return result
```

```python
# Sprawdź działanie funkcji dla słownika: {'Industry':'Factory Under Newspaper'}
Buildshortcut({'Industry':'Factory Under Newspaper'})
```

```python
# Dodaj do fortune kolumnę IndustryShort, która zawierać będzie skrót wyznaczony
# przez funkcję BuildShortcut dla każdego wiersza. Wyświetl nagłówek fortune.
fortune['IndustryShort'] = fortune.apply(Buildshortcut, axis=1)
fortune.head()
```
