<b>Zadanie 1.</b><br>
Dokonując przekształceń ciągów znaków w każdej ze zmiennych zadeklarowanych na Listing 1.
pozostaw tekst "Jana III Sobieskiego". Wielkość liter ma znaczenie. Do manipulacji użyj wycięć
zakresów (slice), podmian (replace), zmian wielkości liter (lower, upper, title).

```python
# Listing 1

lst = ['UL. JANA III SOBIESKIEGO 1/2',
'ulica Jana III Sobieskiego 1 apt 2',
'os. Jana III Sobieskiego',
'plac Jana III SobieSKiego 1/2',
'aleja JANA III Sobieskiego',
'alei Jana III Sobieskiego 1/2',
'ul. jana III SOBIESKIEGO 1 m. 2',
'os. Jana Iii Sobieskiego 1 apt 2']
```

```python
def clearSobieskiText(lst):    
    text_1 = 'jana'
    text_2 = 'iii'
    text_3 = 'sobieskiego'
    len_3 = len(text_3)

    clear_lst = []
    for item in lst:
        item = item.lower()
        index_start = item.index(text_1)
        index_stop = item.index(text_3) + len_3
        item = item[index_start:index_stop]

        if text_1 in item:    
            item = item.replace(text_1,text_1.title())
        if text_2 in item:
            item = item.replace(text_2,text_2.upper())
        if text_3 in item:        
            item = item.replace(text_3,text_3.title())

        clear_lst.append(item)
    
    return clear_lst
```

```python
clearSobieskiText(lst)
```

```python
def getUniqueKeys(data):

    set_keys = set()
    lst_keys = []

    for item in data:
        lst_keys.append(list(item.keys()))    # konwersja dict_keys do listy

    for lst in lst_keys:
        for elem in lst:
            set_keys.add(elem)
    
    return set_keys
```

<b>Zadanie 5B.</b><br>
Napisz metody total() oraz average(), które będą wyliczały odpowiednio sumę lub średnią z
wszystkich pomiarów sepal_length, sepal_width, petal_length, petal_width dla danego obiektu.

```python
import numpy as np

class Iris:
    def __init__(self, sepal_length, sepal_width, petal_length, petal_width):
        self.sepal_length = sepal_length
        self.sepal_width = sepal_width
        self.petal_length = petal_length
        self.petal_width = petal_width
        
    def total(self):
        data = [self.sepal_length, self.sepal_width, self.petal_length, self.petal_width]
        total_sum = np.sum(data)
        return total_sum
    
    def average(self):
        data = [self.sepal_length, self.sepal_width, self.petal_length, self.petal_width]
        average = np.mean(data)
        return average
    
            
class Virginica(Iris):
    def __init__(self, sepal_length, sepal_width, petal_length, petal_width):
        super().__init__(sepal_length, sepal_width, petal_length, petal_width)
        self.species = 'virginica'
        
class Versicolor(Iris):
    def __init__(self, sepal_length, sepal_width, petal_length, petal_width):
        super().__init__(sepal_length, sepal_width, petal_length, petal_width)
        self.species = 'versicolor'
        
class Setosa(Iris):
    def __init__(self, sepal_length, sepal_width, petal_length, petal_width):
        super().__init__(sepal_length, sepal_width, petal_length, petal_width)
        self.species = 'setosa'
```

```python
my_test_iris = Setosa(5.8,9.7,5.1,8.9,)

print(my_test_iris.species)
print(my_test_iris.sepal_length)

print(my_test_iris.average())
print(my_test_iris)
```

<b>Zadanie 5C.</b><br>
Iterując po DATA z Listing 5. twórz obiekty klasy odpowiedniej dla nazwy gatunku (ostatni
rekord każdej z krotek). Obiekt inicjalizuj danymi z pomiarów. Na ekranie wyświetlaj nazwę
gatunku oraz sumę i średnią z pomiarów.

```python
# Listing 5
DATA = [
('Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Species'),
(5.8, 2.7, 5.1, 1.9, 'virginica'),
(5.1, 3.5, 1.4, 0.2, 'setosa'),
(5.7, 2.8, 4.1, 1.3, 'versicolor'),
(6.3, 2.9, 5.6, 1.8, 'virginica'),
(6.4, 3.2, 4.5, 1.5, 'versicolor'),
(4.7, 3.2, 1.3, 0.2, 'setosa'),
(7.0, 3.2, 4.7, 1.4, 'versicolor'),
(7.6, 3.0, 6.6, 2.1, 'virginica'),
(4.9, 3.0, 1.4, 0.2, 'setosa'),
(4.6, 3.1, 1.5, 0.2, 'setosa'),
]
```

```python
# stworzenie listy pomiarow
def getMeasure(data):
    rows_num = len(data)
    cols_num = len(data[0])
    measure_lst = []
    measure = ''
    
    for i in range(1,rows_num):

        for j in range(cols_num-1):
            measure = data[i][:-1]
        measure_lst.append(measure)
        measure = ''  
    return measure_lst
```

```python
# stworzenie listy gatunkow
def getSpecies(data):
    size = len(data)
    cols = len(data[0])
    species_lst = []
    
    for i in range(1,size):
        species_lst.append(data[i][cols-1])
    return species_lst
```

```python
# tworzenie nowych obiektow irysow
def createIrisObject(data):
    measureLst = getMeasure(data)
    speciesLst = getSpecies(data)
    irisLst = []
        
    for i in range(len(data)-1):
        if speciesLst[i] == 'virginica':
            new_iris = Virginica(measureLst[i][0],measureLst[i][1],measureLst[i][2],measureLst[i][3])
                    
        elif speciesLst[i] == 'versicolor':
            new_iris = Versicolor(measureLst[i][0],measureLst[i][1],measureLst[i][2],measureLst[i][3])
        
        elif speciesLst[i] == 'setosa':
            new_iris = Setosa(measureLst[i][0],measureLst[i][1],measureLst[i][2],measureLst[i][3])
        
        irisLst.append(new_iris)
            
    return irisLst
```

```python
new_irises = createIrisObject(DATA)
```

```python
print(new_irises[5].sepal_length)
print(new_irises[5].average())
```

```python
def irisStats(obj):

    stats_lst = []
    i = 0
    
    for item in obj:

        tot = round(obj[i].total() ,3)
        ave = round(obj[i].average() ,3)
        spe = obj[i].species
        stats_lst.append((spe, tot, ave))
        i += 1
    
    return stats_lst
```

```python
irisStats(new_irises)
```

<b>Zadanie 6A.</b><br>
Wykorzystując numpy dla ziarna losowości równego 0 wygeneruj macierz o nazwie A, 50x50
losowych liczb całkowitych z zakresu od 0 do 1024 włącznie. Stwórz macierz B, która będzie
zawierała liczby z macierzy A będące potęgami dwójki. Pozostaw tylko i wyłącznie unikalne
wartości. Uporządkuj macierz B w kolejności malejącej (od największej do najmniejszej).

```python
import numpy as np
import math

N = 50
MIN = 0
MAX = 1024
np.random.seed(seed=0)
a = np.random.randint(MIN, MAX+1, N**2).reshape(N,N)
a
```

```python
def isPowerOf2(num):
    if num != 0:
        num_log = math.log2(num)
        if num_log%2 == 0 or num_log%2 == 1:
            result = True
        else:
            result = False
    else:
        result = False
    return result
```

```python
def onlyPowerOf2(array):
    array_set = set()
    for i in range(N):
        for j in range(N):
            isPower = isPowerOf2(array[i,j])
            if isPower == True:
                array_set.add(array[i,j])
    b = np.array(list(array_set))
    return b        
```

```python
b = onlyPowerOf2(a)
np.sort(b)
```

Sepal length, Sepal width, Petal length, Petal width, Species i wylicz statystyki opisowe dla

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

r = requests.get("https://github.com/AstroMatt/book-python/blob/master/numerical-analysis/data/iris-dirty.csv")
page_soup = r.text
soup = BeautifulSoup(page_soup, features='html.parser')
page_contain = soup.find_all("tr", {'class' : 'js-file-line'})
```

```python
len(page_contain)
```

```python
for item in page_contain:
    print(item)
```

```python
header = soup.select('tr > th')
containers = soup.select('tr > td')
containers[:12]
```

```python
containers = containers[1:]
containers[:12]
```

```python
size = len(page_contain)
lst_all = []
i = 0
for item in range(size-1):    
    sepal_length = containers[i+1].text
    sepal_width = containers[i+2].text
    petal_length = containers[i+3].text
    petal_width = containers[i+4].text
    species = containers[i+5].text
    lst_all.append([sepal_length, sepal_width, petal_length, petal_width, species])
    i += 6
```

```python
len(lst_all)
```

```python
df = pd.DataFrame(lst_all, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])
df.head()
```

```python
df['sepal_length'] = df['sepal_length'].astype('float')
df['sepal_width'] = df['sepal_width'].astype('float')
df['petal_length'] = df['petal_length'].astype('float')
df['petal_width'] = df['petal_width'].astype('float')
```
