```python
import re

def change_exp_level(level):
    pass

def change_exp_skills(level):
    pass

def convert_salary_string(salary_str):
    pass

def go_to_offer(idx):
    element = driver.find_elements_by_css_selector("a.item")
    element[idx].click()
    
def get_offer_data():
    
#     title = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > offer-header > div > div.background.java > div > div.offer-essential-info > div.offer-title-row > span"
#     adress = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > offer-header > div > div.background.java > div > div.offer-essential-info > div.offer-address"
#     company_name = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > offer-header > div > div.tiles > div.tile.company-name > div.value > a"
#     company_size = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > offer-header > div > div.tiles > div.tile.company-size > div.value"
#     employment_type = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > offer-header > div > div.tiles > div.emp-type.tile.permanent > div.value"
#     experience_level = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > offer-header > div > div.tiles > div.tile.exp-level > div.value"
#     added = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > offer-header > div > div.tiles > div.tile.age > div.value"
#     skill_java = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > div > div.card-section.skill-section > div.card-content > div > div:nth-child(1) > div > div.level-title"
#     skill_javascript = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > div > div.card-section.skill-section > div.card-content > div > div:nth-child(2) > div > div.level-title"
#     skill_spring = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > div > div.card-section.skill-section > div.card-content > div > div:nth-child(3) > div > div.level-title"
#     skill_vuejs = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > div > div.card-section.skill-section > div.card-content > div > div:nth-child(4) > div > div.level-title"
#     skill_kotlin = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > div > div.card-section.skill-section > div.card-content > div > div:nth-child(5) > div > div.level-title"
#     description = "#body-top > div.main-content > div.sidebar.offers > div > ui-view > div.offer-details > div > div:nth-child(3) > div.card-content.description"
    
    salary = ""
    while salary == "":
        salary = driver.find_elements_by_css_selector("div.salary-row")[0].find_elements_by_class_name("salary")[0].text
    
    title2 = driver.find_element_by_xpath('//*[@id="body-top"]/div[2]/div[2]/div/ui-view/div[2]/offer-header/div/div[1]/div/div[2]/div[2]/span')
    adress = driver.find_element_by_xpath('//*[@id="body-top"]/div[2]/div[2]/div/ui-view/div[2]/offer-header/div/div[1]/div/div[2]/div[3]')
    company_name = driver.find_element_by_xpath('//*[@id="body-top"]/div[2]/div[2]/div/ui-view/div[2]/offer-header/div/div[2]/div[1]/div[2]/a')
    company_size = driver.find_element_by_xpath('//*[@id="body-top"]/div[2]/div[2]/div/ui-view/div[2]/offer-header/div/div[2]/div[2]/div[2]')
    employment_type = driver.find_element_by_xpath('//*[@id="body-top"]/div[2]/div[2]/div/ui-view/div[2]/offer-header/div/div[2]/div[3]/div[2]')
    experience_level = driver.find_element_by_xpath('//*[@id="body-top"]/div[2]/div[2]/div/ui-view/div[2]/offer-header/div/div[2]/div[4]/div[2]')
    added = driver.find_element_by_xpath('//*[@id="body-top"]/div[2]/div[2]/div/ui-view/div[2]/offer-header/div/div[2]/div[5]/div[2]')
    description = driver.find_element_by_xpath('//*[@id="body-top"]/div[2]/div[2]/div/ui-view/div[2]/div/div[3]/div[2]/div[1]/div[1]')
    
#     adress = driver.find_elements_by_css_selector(adress).text
#     company_name = driver.find_elements_by_css_selector(company_name).text
#     company_size = driver.find_elements_by_css_selector(company_size).text
#     employment_type = driver.find_elements_by_css_selector(employment_type).text
#     experience_level = driver.find_elements_by_css_selector(experience_level).text
#     added = driver.find_elements_by_css_selector(added).text
#     description = driver.find_elements_by_css_selector(description).text
    
#     print(salary, title, adress, company_name, company_size, employment_type, experience_level, added, description)

    print(salary, title2, adress, company_name, company_size, employment_type, experience_level, added, description)
    

    
# od Zbyszka:
# def read_offer():
#     selectors = {
#         "salary": "div.offer-details div.salary-row > span",
#         "title": "div.offer-details div.offer-title-row > span",
#         "address": "div.offer-details div.offer-address",
#         "company_name": "div.offer-details div.tile.company-name > div.value > a",
#         "company_size": "div.offer-details div.tile.company-size > div.value",
#         "employment_type": "div.offer-details div.emp-type.tile > div.value",
#         "experience_level": "div.offer-details div.tile.exp-level > div.value",
#         "age": "div.offer-details div.tile.age > div.value",
#         "description": "div.offer-details > div > div:nth-child(3) > div.card-content.description > div.trix-content"
#     }

#     skills_selector = "div.offer-details div.card-section.skill-section > div.card-content > div"
#     company = ""

#     while company == "":
#         # spowolnić timeoutem
#         company = driver.find_elements_by_css_selector("div.tile.company-name")[0].find_elements_by_class_name('value')[0].text

#     data = {}
#     for k in selectors:
#             element = driver.find_element_by_css_selector(selectors[k])
#             data[k] = element.text

#     skills = driver.find_element_by_css_selector(skills_selector).text.split("\n")
#     for i in range(len(skills)//2):
#         data[skills[i*2]] = skills[i*2+1]

#     print(data)
#     return data

def change_exp_skills(level):
    if level == "nice to have":
        return 1
    if level == "junior":
        return 2
    if level == "regular":
        return 3
    if level == "advanced":
        return 4
    if level == "master":
        return 5
    
def change_exp_level(level):
    if level == 'Junior':
        return 1
    if level == 'Mid':
        return 2
    if level == 'Senior':
        return 3
    
def convert_sallary_string(sallary_str):
    sallary_rex = re.compile('([\d+ \.,]+) - ([\d+ \.,]+) (\w+) ([\w+\/]+)')
    match = sallary_rex.match(sallary_str)

    sallary_from = match[1]
    sallary_to = match[2] 
    sallary_currency = match[3]
    sallary_type = match[4]
    sallary_from = float(sallary_from.replace(' ', ''))
    sallary_to = float(sallary_to.replace(' ', ''))

    if sallary_currency == 'USD' :
        sallary_from = sallary_from * 3.8
        sallary_to = sallary_to * 3.8
        sallary_currency = 'PLN'

    elif sallary_currency == 'CHF' :
        sallary_from = sallary_from * 3.7
        sallary_to = sallary_to * 3.7
        sallary_currency = 'PLN'

    elif sallary_currency == 'EUR' :
        sallary_from = sallary_from * 4.2
        sallary_to = sallary_to * 4.2
        sallary_currency = 'PLN'

    if "gross/month" == sallary_type:
        sallary_from = sallary_from * 0.7
        sallary_to = sallary_to * 0.7

    return sallary_from, sallary_to

def read_offer():
    company = ""
    while company == "":
        # spowolnić timeoutem
        print(driver.find_elements_by_css_selector("div.tile.company-name")[0])
        company = driver.find_elements_by_css_selector("div.tile.company-name")[0].find_elements_by_class_name('value')[0].text
        
        
    title = driver.find_elements_by_css_selector("div.offer-title-row")[0].find_elements_by_class_name("title")[0].text
    emp_type = driver.find_elements_by_css_selector("div.emp-type.tile")[0].find_elements_by_class_name('value')[0].text
    exp_lvl = driver.find_elements_by_css_selector("div.tile.exp-level")[0].find_elements_by_class_name('value')[0].text
    exp_lvl = change_exp_level(exp_lvl)
    salary = driver.find_elements_by_css_selector("div.salary-row")[0].find_elements_by_class_name("salary")[0].text
    salary_min, salary_max = convert_sallary_string(salary)
    adres = driver.find_elements_by_css_selector("div.offer-address")[0].text
    city = adres[-adres[::-1].find(' '):]
    skills_dict= {
        'company': company,
        'emp_type': emp_type,
        'exp_lvl': exp_lvl,
        'title': title,
        'salary_min': salary_min,
        'salary_max': salary_max,
        'city': city,
   }

    skills=driver.find_elements_by_class_name('skill-wrapper')
    for element in skills:
        name=element.find_element_by_class_name('name')
        name_text=name.text
        company_level = element.find_element_by_class_name('level-title')
        level_text=change_exp_skills(company_level.text)
        skills_dict[name_text]=level_text

    return skills_dict

    
def back_to_list():
    back_button = "i.material-icons.waves-effect.waves.back"
    element = driver.find_element_by_css_selector(back_button)
    element.click()

def scroll_to_offer():
    driver.execute_script("""
    const offset = window.pageYOffset
    window.scrollTo(0, offset + 96)   
    """)
    #fragment w javascript bo selenium nie obsługuje takiego scrollowania w pythonie
    
```

```python
from selenium import webdriver
driver = webdriver.Chrome("./chromedriver")
driver.get("https://justjoin.it/all/all")
driver.set_window_size(800,1080)
driver.implicitly_wait(10)
offers = []

for i in range(5):
    go_to_offer(i)
    offers.append(read_offer())
    back_to_list()
    scroll_to_offer()

print(offers)
driver.close()
```

```python
import pandas as pd
df_offers = pd.DataFrame(offers)
df_offers.to_csv("offers2.csv")
```
